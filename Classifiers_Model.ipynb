```python
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
url = 'https://drive.google.com/file/d/1Aj1w_zoiIqkUwhdvh3Dl3kd0NGumnrLL/view?usp=sharing'
path = 'https://drive.google.com/uc?export=download&id='+url.split('/')[-2]
Trump_approval = pd.read_csv(path)
```

```python
Trump_approval.head()
```

```python
#Breaking down the data set into two parts
#Features
features=Trump_approval.iloc[:,:-1]

#Target
target=Trump_approval.iloc[:,-1]
```

```python
#View the data two parts
print(features.head())
```

```python
print(target.head())
```

```python
#Spliting the data set
trainX,testX,trainY,testY=train_test_split(features,target,test_size=0.30,shuffle=False,random_state=4)
```

```python
model=KNeighborsClassifier(n_neighbors=2)
```

```python
model.fit(trainX,trainY)
```

```python
#testing model
results=model.predict(testX)
```

```python
print("This is forcasted results {}".format(results))
print("This is gold results {} \n".format(list(testY)))
```

```python
#Evalution
accu_results=accuracy_score(testY,results)
```

```python
print(accu_results*100)
```
