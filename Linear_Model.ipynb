```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
```

```python
url = 'https://drive.google.com/file/d/16Vl_yMLHOPqZAPUQRUlpp0gZEYmzgHWQ/view?usp=sharing'
path = 'https://drive.google.com/uc?export=download&id='+url.split('/')[-2]
Pollstar_ratings = pd.read_csv(path)
```

```python
print(Pollstar_ratings.head())
```

```python
print(Pollstar_ratings.tail())
```

```python
Pollstar_ratings = Pollstar_ratings.dropna()
```

```python
Null_Columns = Pollstar_ratings.isnull().sum()
print("Total Number of Null Values by Columns in the Dataset")
print(Null_Columns)
```

```python
Features = Pollstar_ratings.iloc[:,:-1]
```

```python
print(Features.head())
```

```python
print(Features.tail())
```

```python
Target=Pollstar_ratings.iloc[:,-1]
```

```python
print(Target.head())
```

```python
trainX,testx,trainY,testy=train_test_split(Features,Target,test_size=0.30,random_state=4,shuffle=False)
```

```python
print(len(trainX))
print(len(trainY))
```

```python
print(len(testx))
print(len(testy))
```

```python
model=LinearRegression()
model.fit(trainX,trainY)
```

```python
result=model.predict(testx)
```
